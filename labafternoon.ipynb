```python
url4 = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
```

```python
response4 = requests.get(url4)
response4.status_code
```

```python
soup4 = BeautifulSoup(response4.content, "html.parser")
```

```python
soup4
```

```python
soup4.select('table.wikitable.sortable.static-row-numbers.jquery-tablesorter')
```

```python
row_count = 0
language = []
numbers = []

if response4.status_code == 200:
    soup4 = BeautifulSoup(response4.text, 'html.parser')

    table = soup4.find('table', class_='wikitable')
    
    for row in table.find_all('tr')[1:]:
        columns = row.find_all(['td','th'])
        language.append(columns[0].text.strip())
        numbers.append(columns[1].text.strip())
        
        row_count+=1
        if row_count >= 10:
            break
    
    languages = pd.DataFrame({'Language':language, 'Native Speakers':numbers})
    

else:
    print(f"Failed to retrieve the page. Status code: {response4.status_code}")
```
